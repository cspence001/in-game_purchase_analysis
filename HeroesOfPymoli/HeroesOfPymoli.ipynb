## Purchasing Analysis (Total)

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#groupby SN
top_spenders = purchase_data.groupby(["SN"])
top_spenders.count().head()

#variable calculations
purchase_total3 = top_spenders["Price"].sum()
purchase_count3 = top_spenders["Purchase ID"].count()
average_purchase3 = (purchase_total3/purchase_count3).map("${:.2f}".format)

#dataframe
top_spenders = pd.DataFrame({"Purchase Count": purchase_count3,
                            "Average Purchase Price": average_purchase3,
                             "Total Purchase Value": purchase_total3
                                  })

#sort based on totals
sorted_spenders = top_spenders.sort_values("Total Purchase Value", ascending=False)
sorted_spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
most_popular = purchase_data[["Item ID", "Item Name", "Price"]]
most_popular.head()
grouped= most_popular.groupby(["Item ID","Item Name"])
grouped.head()


purchase_count = grouped["Item ID"].count()
item_price = grouped["Price"].mean()
total_purchase_value = purchase_count * item_price

#create DataFrame

most_popular_items = pd.DataFrame({"Purchase Count" : purchase_count,
                                    "Item Price" : item_price,
                                    "Total Purchase Value" : total_purchase_value})
 

popular_sorted = most_popular_items.sort_values("Purchase Count", ascending=False)
popular_sorted.head()
```
